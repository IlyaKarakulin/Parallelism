```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("res.csv")
data
```

```python
val_list_40k = data.iloc[0].values.tolist()
boost_list_40k = [val_list_40k[0] / x for x in val_list_40k]

count_thrd = [int(x) for x in data.columns.values.tolist()]
```

```python
plt.figure(figsize=(6, 6))
plt.plot(count_thrd, boost_list_40k, label='n = 100.000')
plt.xticks(count_thrd)
plt.grid(True)
plt.legend()
plt.savefig(f'./times.png', dpi=300, bbox_inches='tight')
plt.show()
```
